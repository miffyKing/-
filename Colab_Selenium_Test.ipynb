```python
# -*- coding: UTF-8 -*-
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
#Colab에선 웹브라우저 창이 뜨지 않으므로 별도 설정한다.
 
options = webdriver.ChromeOptions()
options.add_argument('--headless')        # Head-less 설정
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver', options=options)
```

```python
#해당 url로 이동
url = "https://www.bandtrass.or.kr/customs/total.do?command=CUS00301_POPDetail&SELECT_DIV1=LOCATION_DIV&GODS_TYPE=H&ECONO_TYPE=undefined&PORT_TYPE=undefined&LOCATION_TYPE=A&FILTER1_GODS_UNIT=&SELECT_DIV2=GODS_DIV&FILTER2_GODS_UNIT=6&SELECT_DIV3=&FILTER3_GODS_UNIT=&POP_TABLE=&COL_NAME=&EXCEL_LOG=&MENU_CODE=CUS00301_POP&EXCEL_SUBJECT=&SelectCd1=대구&SelectCd2=284190&S0010003=M" 
driver.get(url)
driver.implicitly_wait(2)

```
